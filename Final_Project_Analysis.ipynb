```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

microsoftDF = pd.read_csv("/content/drive/MyDrive/IT 496/IT496 Microsoft.csv")
appleDF = pd.read_csv("/content/drive/MyDrive/IT 496/IT496 Apple.csv")
vulnerabilityDF = pd.read_csv("/content/drive/MyDrive/IT 496/Final Project/IT496 Final Project.csv")
vendorsProductDF = pd.read_csv("/content/drive/MyDrive/IT 496/IT496 Vendor Product.csv")
```

```python
#Draw a pie chart of top 5 highest total vulnerabilites
#import matplotlib for graphs
#import matplotlib.pyplot as plt
pieName = vulnerabilityDF['Vendor Name'].head(5)
pieValues = vulnerabilityDF['Number of Total Vulnerabilities'].head(5).astype(int)

plt.pie(pieValues, labels=pieName,
        startangle = 90)
plt.show()
```

```python
appleIndex = appleDF.iloc[0,2:].index
appleData = appleDF.iloc[24,2:].values

plt.barh(appleIndex, appleData)
```

```python
microsoftIndex = microsoftDF.iloc[0,2:].index
microsoftData = microsoftDF.iloc[24,2:].values

plt.barh(microsoftIndex, microsoftData)
```

```python
barName = vulnerabilityDF.iloc[0,3:13].index
microsoftVulScore = vulnerabilityDF.iloc[0,3:13].values
appleVulScore = vulnerabilityDF.iloc[4,3:13].values

df = pd.DataFrame({'Apple Vulnerabilities Score': appleVulScore, 'Microsoft Vulnerabilities Scores': microsoftVulScore}, index=barName)

# plot
ax = df.plot(kind='bar', figsize=(6, 4), rot=0,)
plt.xticks(rotation=90)

plt.xlabel("Vulnerability Score")
plt.ylabel("No. of vulnerabilites")
plt.title("Number of Vulnerability Scores of Microsoft and Apple")
plt.show()
```

```python
appleIndex = appleDF.iloc[0,2:].index
appleData = appleDF.iloc[24,2:].values
microsoftData = microsoftDF.iloc[24,2:].values

# create the dataframe
df = pd.DataFrame({'Apple Vulnerabilities': appleData, 'Microsoft Vulnerabilities': microsoftData}, index=appleIndex)

# plot
ax = df.plot(kind='bar', figsize=(6, 4), rot=0, title='Number of Vulnerability Comparison by Type', ylabel='Number of Vulnerabilities')
plt.xticks(rotation=90)
plt.show()
```

```python
Year = appleDF.iloc[0:24,0].values
appleVulYear = appleDF.iloc[0:24,1].values
microsoftVulYear = microsoftDF.iloc[0:24,1].values

# create the dataframe
df = pd.DataFrame({'Apple Vulnerabilities by Year': appleVulYear, 'Microsoft Vulnerabilities': microsoftVulYear}, index=Year)

# plot
ax = df.plot(figsize=(6, 4), rot=0,)
plt.xticks(rotation=90)

plt.xlabel("Year")
plt.ylabel("No. of vulnerabilites")
plt.title("Number of Vulnerability by Year of Microsoft and Apple")
plt.show()
```

```python
microsoftVulNumbers = vendorsProductDF['Number of Products'][0]
appleVulNumbers = vendorsProductDF['Number of Products'][4]
microsoftProducts = vendorsProductDF['Number of Vulnerabilities'][0]
appleProducts = vendorsProductDF['Number of Vulnerabilities'][4]

microsoftRatio = microsoftProducts/microsoftVulNumbers
appleRatio = appleProducts/appleVulNumbers

ratioValue = [microsoftRatio, appleRatio]
xValues = ['Microsoft', 'Apple']

df = pd.DataFrame({'ratio Values': ratioValue}, index=xValues)

# plot
ax = df.plot(kind='bar')
plt.ylim(0,60)
plt.xticks(rotation=0)
plt.xlabel("Company")
plt.ylabel("Number of vulnerabiltiy Value")
plt.title("Number of Vulnerabilites per Products")
plt.show()
```

```python
percentageValue = [microsoftRatio, appleRatio]
explode = (0, 0.1)

plt.pie(percentageValue, labels=['Microsoft Vulnerability %','Apple Vulnerability %'],
        startangle = 90, explode = explode, autopct='%1.1f%%')
plt.show()
```
