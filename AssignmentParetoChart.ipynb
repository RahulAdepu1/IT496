```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

#upload the HelpDesk.csv to your Google Drive
df = pd.read_csv("/content/drive/MyDrive/IT 496/Rahul_Adepu-Pareto_Chart_Assignment.csv")
print(df)
```

```python
#set this column as index - to use as plot labels
df.index = df['Quality']
#descending sort
df = df.sort_values(by = 'Importance', ascending = False)
print(df)
```

```python
#cumulative percentage column is created using cumsum function
df["cumulativePercentage"] = df["Importance"].cumsum()/df["Importance"].sum()*100

df["cumulativePercentage"] = df["cumulativePercentage"].apply(lambda x: round(x,2))
print(df)
```

```python
flg, ax = plt.subplots()
ax.bar(df.index, df["Importance"], color = "C0")

#show labels diagonally
ax.set_xticklabels( df['Quality'], rotation = 90)

#ax2 and ax1 wil be shown at the same time with twinx
ax2 = ax.twinx()

ax2.plot(df.index, df["cumulativePercentage"], color = "C1", marker="D", ms = 7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis = "y", colors = "C0")
ax2.tick_params(axis = "y", colors = "C1")

plt.show()
```
